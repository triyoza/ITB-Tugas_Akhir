## Annual maxima

```python
maksimum = {}
for i,j in enumerate (namastat):
    max_tahun = np.array([])
    for k in range(1,42):
        x = data[i].loc[data[i]['Th'] == k, ['RR']]
        x1 = np.max(x)
        max_tahun = np.append(max_tahun, x1)
    maksimum[str(j)] = max_tahun
```

```python
maksimum_edit = {}
for stas in maksimum:
    maksimum_edit[stas] = [i for i in (maksimum[stas][~np.isnan(maksimum[stas])])]
#maksimum_edit
```

```python
maksimum_20th = {}
for i in maksimum_edit:
    if len(maksimum_edit[i]) > 19:
        maksimum_20th[i] = maksimum_edit[i]
```

```python
#Peerhitungan nilai curah hujan ekstrem periode ulang untuk semua stasiun 
T = [2, 5, 10, 25, 50, 100]
ch_pugb = {}
for i,j in enumerate (maksimum_20th):
    ch_fitgb = stats.gumbel_r.fit(maksimum_20th[j])
    ch_Tgb = np.array([])
    
    for k in range (6):
        periode_ulanggb = stats.gumbel_r.ppf([1-(1/T[k])], * ch_fitgb)
        ch_Tgb=np.append(ch_Tgb,periode_ulanggb)
        #ch_pu[i] = np.append(ch_pu[i],periode_ulang)
    ch_pugb[str(j)] = ch_Tgb
```
